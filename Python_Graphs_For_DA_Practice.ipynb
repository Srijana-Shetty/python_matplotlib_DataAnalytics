```python
plt.plot(x, y, marker='d')
```

```python
x=df['work_year']
y=df['salary']
bar_colors = ['skyblue', 'lightcoral', 'lightgreen', 'gold']
plt.bar(x,y, color=bar_colors)
plt.xlabel('work_year')
plt.ylabel('salary')
plt.title('Bar Plot')
plt.show()
```
