```python
import matplotlib.pyplot as plt
import seaborn as sns
data = sns.load_dataset('titanic')
plt.figure(figsize=(15, 10))
plt.subplot(3,3,1)
sns.scatterplot(x='age',y='pclass',data=data,hue='class')
plt.subplot(3,3,2)
sns.boxplot(x='class',y='age',data=data)
plt.subplot(3,3,3)
sns.barplot(x='class',y='age',data=data, hue='sex')
plt.subplot(3,3,4)
sns.countplot(x='class',data=data)
plt.subplot(3,3,5)
sns.stripplot(x='age',y='class',data=data)
plt.subplot(3,3,6)
sns.lineplot(x='age',y='class',data=data)
plt.show()
```
